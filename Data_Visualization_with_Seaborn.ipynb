Seaborn - It is a visualization library in Python which is based on Numpy.

It is used to plot 2D & 3D graphs.

Used for Data Viz.

Used to convert numerical data into graphical form.

```
import seaborn as sns
import pandas as pd   
```

```
#load a dataset in seaborn
df = sns.load_dataset('iris')
```

1.Joint Plot

2.Dist Plot

3.Pair Plot



```
df.corr()
```

```
#corr with heatmap
#heatmap - it is colored cell, typically in monochromatic 
#scale to show 2d corr. matrix between two discrete dimensions.
sns.heatmap(df.corr())
```

```
#joint plot
# A joint plot allow to study the relationship between 
#2numeric variable.
#It disply the corr. It is usually scatter,hex,2d or density plot
```

```
df.columns
```

```
sns.jointplot(x='tip',y='total_bill',data=df )
```

```
sns.jointplot(x='tip',y='total_bill',data=df,kind ='hex' )
```

```
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg' )
```

```
#task - perform a joinplot with x= tip y = size
#pair plot - It is known as scatter plot in which one variable
#in the same data row is matched to another variable.

sns.pairplot(df)
```

```
sns.pairplot(df,hue='smoker')
```

```
#dist plot - Distribution plot
# It help to check the distribution of column feature,
#also used to create histograms.

sns.distplot(df['total_bill'])
```

```
sns.distplot(df['tip'],kde =False,bins=10)
#kde - kernel density estimation
```

Categorical Plot

1. CountPlot

2. Barplot

3. Violinplot

4. Boxplot

```
#countplot - It shows the count of observation in each 
#bin using bars

sns.countplot('day',data=df)
```

```
sns.countplot('smoker',data=df)
```

```
sns.countplot('sex',data=df)
```

```
#bar plot 
sns.barplot(y='total_bill',x='sex',data=df)
```

```
#boxplot- It is a graph that present information from
#five no. summary
sns.boxplot('sex','total_bill',data=df)
#iqr - Interquartile rate
```

```
#task - plot a graph with boxplot x= day,y = total_bill
```

```
#violin plot - help us to see the distribution of data in terms of
#kde & boxplot.
sns.violinplot(x='total_bill',y='day',data=df)
```
